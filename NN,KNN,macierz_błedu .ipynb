```python
import numpy as np 
import pandas as pd 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# pobranie pliku
df = pd.read_csv("C:/Users/magda/Downloads/leaf/leaf.csv",header=None) 
df 
```

```python
# wyodrębnienie dwóch typów liści i wyświetlenie zbioru
data=df[(df[0]==3)|(df[0]==5)]
data
```

```python
# podział zbioru na dwa pobodnej wielkości podzbiory train(zbiór wykorzystywany przy uczeniu klasyfikatora)
#oraz test (zbiór uzywany przy późniejszej klasyfikacji na wyuczonym juz klasyfikatorze)
train,test=train_test_split(data.copy(),test_size=0.2,random_state=333) 
train 
```

```python
# pobranie pliku
da = pd.read_csv("C:/Users/magda/Downloads/leaf/leaf.csv",header=None) 
da 
```

```python
# wyodrębnienie dwóch typów liści i wyświetlenie zbioru
newdata=da[(da[0]==3)|(da[0]==5)]
newdata
```

```python
# podział zbioru na dwa podzbiory train(zbiór wykorzystywany przy uczeniu klasyfikatora)
#oraz test (zbiór uzywany przy późniejszej klasyfikacji na wyuczonym juz klasyfikatorze). 
#Założenie: mniej próbek w zbiorze treningowym(train2)
train2,test2=train_test_split(newdata.copy(),test_size=0.7,random_state=333) 
train2 
```

```python
test2
```

```python
# wyświetlenie zbiorów przy uzyciu kolorów 
# znaczamy odleglosc na podstawie kolumny 3 i 10
plt.scatter(x=train2[3][train2[0]==3],y=train2[10][train2[0]==3], c="green")
plt.scatter(x=train2[3][train2[0]==5],y=train2[10][train2[0]==5], c="blue") 
plt.scatter(x=test2[3], y=test2[10], c="orange") # te próbki trzeba zaklasyfikowac albo do zielonych albo do niebieskich
```

```python
# funckja obliczajaca odległość 
#A=(x,y) B=(x,y)   
def distance(x_a,y_a,x_b,y_b): 
    return np.sqrt(((x_a-x_b)**2) + ((y_a-y_b)**2))  
# wartość k określająca ilość najbliższych sąsiadów branych pod uwagę 
k=5
```

```python
# tablica służąca do koncowego porównania działania obu klasyfikatorów
result2 = pd.DataFrame(columns=[0,1,2])
#tabela z przewidywanymi wynikami
predicted=[]  
#tabla z prawidłowymi danymi
correct = []
# iterowanie po elementach ze zbioru test
for index_test, row_test in test2.iterrows(): 
    if(row_test[0]==3):
        correct.append("green") 
    else:
        correct.append("blue")
    distances =[] # informacje o dystansach i o klasie
    #iterowanie po elementach ze zbioru train
    for index_train, row_train in train2.iterrows():
        if (row_train[0]==3):
            distances.append([distance(row_test[3], row_test[10], row_train[3], row_train[10]),3])
        else:
            distances.append([distance(row_test[3], row_test[10], row_train[3], row_train[10]),5])
   
    distances.sort(key=lambda distance: distance[0])   
    distances = distances[:k]
#     print("Top 5") 
#     print(distances)  
    #zliczanie elementów konkretnej klasy
    count_3=0; 
    count_5=0; 
    for i in distances[:k]:
        if (i[1]==3): count_3+=1
        else: count_5+=1
    print(count_3," ",count_5)
    if(count_3>count_5): 
        result2.loc[len(result2)] = [row_test[3], row_test[10], 3]
        predicted.append("green")
        if(row_test[0]==3):
            print("Zaklasyfikowano poprawnie")
        else: 
            print("Zaklasyfikowano niepoprawnie")
    else:
        result2.loc[len(result2)] = [row_test[3], row_test[10], 5]
        predicted.append("blue")
        if(row_test[0]==5): 
            print("Zaklasyfikowano poprawnie") 
        else: 
            print("Zaklasyfikowano niepoprawnie") 

            
```

```python
# wykres przedstawia wynik działania drugiego klasyfikatora 
plt.scatter(x=train2[3][train2[0]==3],y=train2[10][train2[0]==3], c="green")
plt.scatter(x=train2[3][train2[0]==5],y=train2[10][train2[0]==5], c="blue") 
plt.scatter(x=result2[0][result2[2]==3],y=result2[1][result2[2]==3], c="green")
plt.scatter(x=result2[0][result2[2]==5],y=result2[1][result2[2]==5], c="blue")
```

```python
#macierz błedu 
cm = confusion_matrix(correct, predicted)
print(cm)
```
